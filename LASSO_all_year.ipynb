```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the data 

full = pd.read_csv('MKTCAP_LN_UNADJ.csv', encoding = 'cp932')
fullr = full[full.columns[1:]]

# remove all zero columns in the complete and correct dataframe 

d = fullr.loc[:, (fullr != 0).any(axis = 0)]
d.head()
```

```python
# remove rows in all years with zero MKTCAP/ the last column

dt = d[d['value'] != 0]

dt.groupby(dt['ddate']).count()
```

```python
# groupby sumamrization from the datset wihtout dropping the zero MKTCAP column

d.groupby(d['ddate']).count()
```

```python
# 2014 and 2015 data, get the overlapping companies from them 

x_14 = dt[dt['ddate'] == 2014][dt.columns[:-1]]
y_15 = dt[dt['ddate'] == 2015][['security_code', 'value']]

# the overlapping part 
data = pd.merge(x_14, y_15, how = 'inner', on = 'security_code')
print (data.shape)
data.head()
```

```python
data['value'].describe()
```

```python
plt.plot(data['value'])
```

```python
# feature list 

feature = data.columns[2:-1]
```

```python
# load the previously pickled model from disk

import pickle 

lasso1516 = pickle.load(open('LN_LASSO_model.sav', 'rb'))

# 15 16 model on predicting 15 values from 14 features - R^2
lasso1516.score(data[feature], data['value'])
```

```python
# 15 16 model on predicting 15 values from 14 features - MSE

SSE = 0
for i in range(data.shape[0]):
    SSE = SSE + (lasso1516.predict(data.iloc[i, :][feature]) - data.iloc[i, :]['value']) ** 2
    
MSE = SSE/ data.shape[0]
MSE
```

```python
MSE
```

```python
# building tunned LASSO on 2014 2015 data 

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(data[feature], data['value'], test_size = 0.2, random_state = 0 )
```

```python
# Cross-validated Lasso, using the LARS algorithm 

from sklearn.linear_model import LassoLarsCV

model = LassoLarsCV(cv = 10, precompute = False).fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.alpha_
```

```python
# CV using R^2 and MSE 

from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score 

MSE = make_scorer(mean_squared_error)


# MSE 10-fold CV
kf = 10
CV_MSE = np.mean(cross_val_score(model, data[feature], data['value'], cv = kf, scoring = MSE))

# R^2 10-fold CV
CV_R2 = np.mean(cross_val_score(model, data[feature], data['value'], cv = kf, scoring = 'r2'))
```

```python
print (CV_MSE)
print (CV_R2)
```
